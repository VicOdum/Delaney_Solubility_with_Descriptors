```python
# import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv(r"C:\Users\USER\Downloads\Data_Analysis_ML\delaney_solubility_with_descriptors.csv")
df
```

```python
df.describe()
```

```python
# Separating the Features from the Target
target = "logS"
X = df.drop(target, axis=1)
y = df[target]
```

```python
# Spliting dataset into train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=100)
```

```python
# Model Building with Linear Regression
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
# Apply model to make predictions on X_train and X_test
y_model_train_pred = model.predict(X_train)
y_model_test_pred = model.predict(X_test)
```

```python
y_model_test_pred
```

```python
# Model Performance Evaluation
y_train
```

```python
y_model_train_pred
```

```python
# Evaluating model performance for mean error
model_train_mse= mean_squared_error(y_train, y_model_train_pred)
model_train_r2 = r2_score(y_train, y_model_train_pred)

model_test_mse= mean_squared_error(y_test, y_model_test_pred)
model_test_r2 = r2_score(y_test, y_model_test_pred)
```

```python
lr_results = pd.DataFrame(["Linear Regression", model_train_mse, model_train_r2, model_test_mse, model_test_r2]).transpose()
```

```python
lr_results.columns= ['METHOD |', 'TRAINING_MSE |', 'TRAINING_R2 |', 'TEST_MSE |', 'TEST_R2']
```

```python
lr_results
```

```python
# Build Model with RandomForest Regression
model_rf = RandomForestRegressor(max_depth=2, random_state=100)
# Train model
model_rf.fit(X_train, y_train)
```

```python
# Apply model to make prediction
y_model_rf_train_pred = model_rf.predict(X_train)
y_model_rf_test_pred = model_rf.predict(X_test)
```

```python
# Evaluate model performance with mean squared error and R2_score
model_rf_train_mse= mean_squared_error(y_train, y_model_rf_train_pred)
model_rf_train_r2 = r2_score(y_train, y_model_rf_train_pred)

model_rf_test_mse= mean_squared_error(y_test, y_model_rf_test_pred)
model_rf_test_r2 = r2_score(y_test, y_model_rf_test_pred)
```

```python
rf_results = pd.DataFrame(["RandomForestRegressor", model_rf_train_mse, model_rf_train_r2, model_rf_test_mse, model_rf_test_r2]).transpose()
```

```python
rf_results.columns = ['METHOD |', 'TRAINING_MSE |', 'TRAINING_R2 |', 'TEST_MSE |', 'TEST_R2']
```

```python
rf_results
```

```python
# Compare the LinearRegression Model and the RandomForestRegression Model
models = pd.concat([lr_results, rf_results], axis=0)
models
```
